```python
# first neural network with keras tutorial
from numpy import loadtxt
from keras.models import Sequential
from keras.layers import Dense , Dropout
import numpy as np
```

```python
# mounting the Google drive
#from google.colab import drive

#drive.mount('/content/drive')

# Load the dataset
fp = "sample_data/datasets_heart_noheader.csv"
```

```python
# load the dataset
dataset = loadtxt(fp, delimiter=',')
# split into input (X) and output (y) variables
print(dataset.shape)
X = dataset[0:270,0:13]
print(X.shape)
y = dataset[0:270,13]
X2 = dataset[270:302,0:13]
y2 = dataset[270:302,13]
print(X2)
print(y2)
```

```python
# define the keras model
model = Sequential()
model.add(Dense(12, input_dim=13, activation='relu'))
model.add(Dense(4, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(4, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(4, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
# compile the keras model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
# fit the keras model on the dataset
model.fit(X, y, epochs=600, batch_size=10)
```

```python
# evaluate the keras model
_, accuracy = model.evaluate(X, y)
print('Accuracy: %.2f' % (accuracy*100))
```

```python
# make probability predictions with the model
predictions = model.predict(X)
# round predictions
rounded = [round(x[0]) for x in predictions]
```

```python
# make class predictions with the model
for i in range(15):
	print('%s => %d (expected %d)' % (X[i].tolist(), rounded[i], y[i]))
```

```python
M= len(rounded)
print( 'Accuracy= ' , 1- np.sum(np.abs( rounded - y )) / M)
```

```python
# evaluate the keras model, testing
_, accuracy = model.evaluate(X2, y2)
print('Accuracy: %.2f' % (accuracy*100))
```

```python
# make probability predictions with the model, testing
predictions = model.predict(X2)
# round predictions
rounded = [round(x[0]) for x in predictions]
```

```python
# make class predictions with the model, testing
for i in range(25):
	print('%s => %d (expected %d)' % (X2[i].tolist(), rounded[i], y2[i]))
```

```python
M= len(rounded)
print( 'Accuracy= ' , 1- np.sum(np.abs(rounded - y2)) / M)
```

1. Run the program a few times. Is there any meaningful difference between the results? Why?
  - Yes, even if the neural network (NN) architecture is not changed and the dataset is exactly the same, running the training process again can and often will change the model performance.
While the evaluation of a pre-trained, frozen model on the same data should produce identical results, the training process itself is stochastic (non-deterministic).
Here is a breakdown of why this happens and how to manage it:
Why Results Change (Sources of Randomness)
Random Weight Initialization: Neural networks start by initializing their internal weights with random numbers. Different initializations lead to different paths during optimization, resulting in different final models.
Data Shuffling: Data is usually shuffled at the start of each epoch to improve learning. If not strictly controlled, this means the model sees examples in a different order each time, affecting weight updates.
GPU Non-determinism: Many GPU operations (e.g., in CUDA/PyTorch/TensorFlow) are parallelized and not strictly deterministic. The order of operations may vary slightly, leading to different floating-point results that accumulate over training.
Hardware/Numerical Differences: Small differences in how CPUs or GPUs handle floating-point math can result in different, yet equally valid, performance metrics.
