Variables/Columns:

*   Height in Cm
*   Weight in Kg
*   SEx, 1= Female, 2= Male
*   Weight in pounds
*   Weight in pounds plus noise
*   random numbers

Then we compute the correlation between variables








```python
# libraries and getting connected to Google drive
# file_path is the file folder in Google Drive plus the dataset file name
import pandas as pd
import numpy as np

# Load the dataset
file_path = "sample_data/celebs2_corr.xlsx"
```

```python
# Update this to your file's location
df = pd.read_excel(file_path)

# Display the first few rows of the dataset (optional)
print("First few rows of the dataset:")
print(df.head())

# Check for numeric columns (correlation is calculated only for numeric data)
numeric_df = df.select_dtypes(include=[np.number])

# Compute the correlation matrix for numerical variables
correlation_matrix = numeric_df.corr()

# Display the correlation matrix
print("\nCorrelation Matrix:")
print(correlation_matrix)

# Save the correlation matrix to a new Excel file (optional)
correlation_matrix.to_excel('sample_data\correlation_matrix.xlsx', index=True)
print("\nCorrelation matrix saved to 'correlation_matrix.xlsx'.")
```
